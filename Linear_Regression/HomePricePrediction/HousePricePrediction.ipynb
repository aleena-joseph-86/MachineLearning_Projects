Below table represents current home prices in monroe township based on square feet area, new jersey<br/>

<img src='homepricetable.JPG'>

**Problem Statement**:<br/> Given above data build a machine learning model that can predict home prices based on square feet area<br/>
(1) Predict price of a home with area = 3300 sqr ft<br/>
(2) Predict price of a home with area = 5000 sqr ft<br/>
(3) Predict multiple values at a time 

#### Solution

```python
# import the necessary libraries
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
# load the data
df = pd.read_csv('homeprices.csv')
df
```

```python
# Plot area vs price
plt.title('AREA VS PRICE')
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,marker='*',color='red')
plt.show()
```

```python
# Train the model 
model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)
```

```python
# Predict 
print(f'price of a home with area = 3300 sqr ft is {model.predict([[3300]])}')
print(f'price of a home with area = 5000 sqr ft is {model.predict([[5000]])}')
```

```python
# verification
         # y = mx + c 
m = model.coef_ 
c = model.intercept_

x = 3300
print(f'price of a home with area = 3300 sqr ft is {m*x+c}')

x = 5000
print(f'price of a home with area = 3300 sqr ft is {m*x+c}')
```

#### Predict multiple values at a time 

```python
# Load the Data 
data = pd.read_csv('areas.csv')
data.tail(3)
```

```python
# Predict the multiple values 
prices = model.predict(data[['area']])
prices
```

```python
# add it to the data 
data['price'] = prices
data.head(3)
```

```python
# Plot the best fit line
plt.scatter(data.area,data.price,marker='+',color='red')
plt.plot(data.area,data.price)
plt.title('AREA VS PRICE')
plt.xlabel('area')
plt.ylabel('price')
plt.show()
```
