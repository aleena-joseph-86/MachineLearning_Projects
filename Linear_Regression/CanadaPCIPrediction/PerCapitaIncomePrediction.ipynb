**Problem Statement**

Predict canada's per capita income in year 2020 using canada_per_capita_income.csv file build a regression model and predict the per capita income of canadian citizens in year 2020

#### Solution

```python
# import the necessary libraries
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
# Load the data 
df = pd.read_csv('canada_per_capita_income.csv')
df.head(3)
```

```python
# Plot year vs pci 
plt.title('YEAR VS PCI')
plt.xlabel('year')
plt.ylabel('pci')
plt.scatter(df.year,df['per capita income (US$)'],marker='+',color='orange')
plt.show()
```

```python
# Train the model 
model = linear_model.LinearRegression()
model.fit(df[['year']],df['per capita income (US$)'])
```

```python
# Plot the best fil line 
plt.title('YEAR VS PCI')
plt.xlabel('year')
plt.ylabel('pci')
plt.scatter(df.year,df['per capita income (US$)'],marker='+',color='orange')
plt.plot(df.year,model.predict(df[['year']]))
plt.show()
```

```python
# Predict 
print(f'the per capita income of canadian citizens in year 2020 is {model.predict([[2020]])}')
```

```python
# Verification
 # y = mx+c 

m = model.coef_ 
c = model.intercept_
x = 2020

print(f'the per capita income of canadian citizens in year 2020 is {m*x+c}')
```
