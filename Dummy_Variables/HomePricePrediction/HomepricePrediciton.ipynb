#### 1.Import the necessary libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

#### 2.Load the data

```python
data = pd.read_csv('homeprices.csv')
data
```

#### 3.Create the dummy variables

```python
dummies = pd.get_dummies(data.town,dtype=int)
dummies
```

#### 4.Concatenate the original dataframe with the generated dummies

```python
concatenated = pd.concat([data,dummies],axis=1)
concatenated
```

#### 5.Handle the redundant data

```python
final = concatenated.drop(['town','monroe township'],axis='columns')
final
```

#### 6.Train the model

```python
X = final.drop(['price'],axis='columns')
y = final.price

model = linear_model.LinearRegression()
model.fit(X,y)
model.score(X,y)
```
