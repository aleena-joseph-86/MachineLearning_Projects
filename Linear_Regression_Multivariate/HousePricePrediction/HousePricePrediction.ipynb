**Problem Statement**

Below is the table containing home prices in monroe twp, NJ. Here price depends on area (square feet), bed rooms and age of the home (in years). Given these prices we have to predict prices of new homes based on area, bed rooms and age.<br/>
<img src='homeprices.jpg'><br/>
Given these home prices find out price of a home that has,<br/>
3000 sqr ft area, 3 bedrooms, 40 year old<br/>
2500 sqr ft area, 4 bedrooms, 5 year old<br/>

#### Solution

```python
# Import the libraries 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# Load the data 
df = pd.read_csv('homeprices (1).csv') 
```

```python
# Fill the missing data 
median_bedrooms = df.bedrooms.median()

df.fillna(median_bedrooms,inplace=True)
```

```python
# Plot area,bedrooms,age vs price 

plt.subplot(1,3,1)
plt.title('Area vs Price')
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,marker='+',color='red') 

plt.subplot(1,3,2)
plt.title('Bedrooms vs Price')
plt.xlabel('bedroom')
plt.ylabel('price')
plt.scatter(df.bedrooms,df.price,marker='+',color='yellow') 

plt.subplot(1,3,3)
plt.title('Age vs Price')
plt.xlabel('age')
plt.ylabel('price')
plt.scatter(df.age,df.price,marker='+',color='orange')

plt.tight_layout()
plt.show()
```

```python
# Train the model 
model = linear_model.LinearRegression()
model.fit(df[['area','bedrooms','age']],df.price)
```

```python
# Predict 
print(f'price of a home that has,3000 sqr ft area, 3 bedrooms, 40 year old: {model.predict([[3000,3,40]])}')
print(f'price of a home that has,2500 sqr ft area, 4 bedrooms, 5 year old: {model.predict([[32500,4,5]])}')
```

```python
# Verification 
                                # price = m1*area + m2*bedrooms + m3*age + c

m1,m2,m3 = model.coef_
c = model.intercept_

area = 3000
bedrooms = 3
age = 40

price = m1*area + m2*bedrooms + m3*age + c
print(f'price of a home that has,3000 sqr ft area, 3 bedrooms, 40 year old: {price}')
```
